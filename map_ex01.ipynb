```python
import folium
import json

m = folium.Map(location=[36.634445, 127.795277], zoom_start=7)
gfile = open('./data/province-simple.json', mode='r', encoding='utf-8').read()
jsondata = json.loads(gfile)
folium.GeoJson(jsondata, name='province').add_to(m)
folium.Marker(location=[36.634445, 127.795277], popup='지도중심').add_to(m)
m
```

```python
m2 = folium.Map(location=[36.505354, 127.704334], zoom_start=7, tiles='Cartodb Positron')
point_data = json.loads(open('./data/point.json', mode='r', encoding='utf-8').read())

for i in range(len(point_data)):
    folium.Circle(location=point_data[i], radius=50, color='#000000', fill='crimson').add_to(m2)
m2
```

```python
import folium
from folium.plugins import HeatMap

m2 = folium.Map(location=[36.505354, 127.704334], zoom_start=7, tiles='Cartodb Positron')

HeatMap(point_data).add_to(m2)
m2
```
